```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel(
    "../data/online_retail_II.xlsx",
    sheet_name="Year 2010-2011"
)

print(df.shape)
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=["Customer ID"])

print(df.shape)
```

```python
print("Duplicate Rows:", df.duplicated().sum())
```

```python
df = df.drop_duplicates()

print(df.shape)
```

```python
(df["Quantity"] < 0).sum()
```

```python
df = df[df["Quantity"] > 0]

print(df.shape)
```

```python
(df["Price"] <= 0).sum()
```

```python
df = df[df["Price"] > 0]

print(df.shape)
```

```python
df["InvoiceDate"] = pd.to_datetime(
    df["InvoiceDate"]
)
```

```python
df["TotalAmount"] = (
    df["Quantity"] *
    df["Price"]
)
```

```python
df[["Quantity","Price","TotalAmount"]].head()
```

```python
df["Year"] = df["InvoiceDate"].dt.year
```

```python
df["Month"] = df["InvoiceDate"].dt.month
```

```python
df["Day"] = df["InvoiceDate"].dt.day
```

```python
df["Weekday"] = (
    df["InvoiceDate"]
    .dt.day_name()
)
```

```python
df["Quarter"] = (
    df["InvoiceDate"]
    .dt.quarter
)
```

```python
df["CustomerFrequency"] = (
    df.groupby("Customer ID")
      ["Invoice"]
      .transform("count")
)
```

```python
def order_size(amount):

    if amount < 50:
        return "Small"

    elif amount < 200:
        return "Medium"

    else:
        return "Large"

df["OrderSize"] = (
    df["TotalAmount"]
    .apply(order_size)
)
```

```python
df.head()
```

```python
df.to_csv(
    "../data/cleaned_retail.csv",
    index=False
)

print("Dataset Saved")
```
