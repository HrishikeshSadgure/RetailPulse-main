## Dataset Loading

Loading Year 2009-2010 and Year 2010-2011 datasets.

```python
df1 = pd.read_excel(
    "../data/online_retail_II.xlsx",
    sheet_name="Year 2009-2010"
)

df2 = pd.read_excel(
    "../data/online_retail_II.xlsx",
    sheet_name="Year 2010-2011"
)

df = pd.concat([df1, df2])

print("Dataset Loaded Successfully")
```

## Quantity Distribution

Understanding product purchase quantities.

```python
plt.figure(figsize=(10,5))

df["Quantity"].hist(bins=50)

plt.title("Quantity Distribution")

plt.show()
```

## Price Distribution

Analyzing product pricing.

```python
plt.figure(figsize=(10,5))

df["Price"].hist(bins=50)

plt.title("Price Distribution")

plt.show()
```

## Top Countries Analysis

Identifying countries with highest transactions.

```python
top_country = df["Country"].value_counts().head(10)

top_country.plot(
    kind="bar",
    figsize=(10,5)
)

plt.title("Top 10 Countries")

plt.show()
```

## Duplicate Records Analysis

Checking for repeated transactions.

```python
print(
    "Duplicate Rows:",
    df.duplicated().sum()
)
```
